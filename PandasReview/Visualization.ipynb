### Matplotlib Functions and Notation

#### Basic Plotting
- **`plt.plot(x, y)`**: Plots `y` versus `x` as lines and/or markers.
- **`plt.scatter(x, y)`**: Creates a scatter plot of `y` versus `x` with points.
- **`plt.bar(x, height)`**: Creates a bar chart with bars of heights specified in `height` at positions `x`.
- **`plt.hist(data, bins)`**: Plots a histogram of the data with the specified number of `bins`.
- **`plt.pie(x)`**: Creates a pie chart with the sizes of the slices given by `x`.

#### Axes and Labels
- **`plt.xlabel(label)`**: Sets the label for the x-axis.
- **`plt.ylabel(label)`**: Sets the label for the y-axis.
- **`plt.title(title)`**: Sets the title of the plot.
- **`plt.legend()`**: Displays a legend on the plot.

#### Customization and Styling
- **`plt.xlim([min, max])`**: Sets the limits for the x-axis.
- **`plt.ylim([min, max])`**: Sets the limits for the y-axis.
- **`plt.grid(True)`**: Adds a grid to the plot.
- **`plt.style.use('style_name')`**: Applies a predefined style to the plot.
- **`plt.annotate(text, xy)`**: Adds an annotation `text` at a specific point `xy` on the plot.

#### Display and Save
- **`plt.show()`**: Displays the plot.
- **`plt.savefig('filename.png')`**: Saves the current figure to a file.

#### Subplots
- **`plt.subplot(rows, cols, index)`**: Creates a subplot in a grid layout with `rows` and `cols`, and selects the subplot at `index`.
- **`plt.subplots()`**: Creates a figure and a grid of subplots.

#### Advanced Plotting
- **`plt.errorbar(x, y, yerr)`**: Plots `y` versus `x` with error bars.
- **`plt.fill_between(x, y1, y2)`**: Fills the area between two curves or a curve and a baseline.
- **`plt.contour(X, Y, Z)`**: Creates a contour plot.
- **`plt.imshow()`**: Displays data as an image.


```python
### Sample Problem
''' 
You are given a list of stock closing prices, prices, and the corresponding dates, dates. Your task is to plot the stock prices along with two moving averages:
    1. A short-term moving average (SMA) using a window size of 5.
    2. A long-term moving average (LMA) using a window size of 20.
'''
dates = [
    '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', 
    '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10',
    '2023-01-11', '2023-01-12', '2023-01-13', '2023-01-14', '2023-01-15', 
    '2023-01-16', '2023-01-17', '2023-01-18', '2023-01-19', '2023-01-20',
    '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24', '2023-01-25',
    '2023-01-26', '2023-01-27', '2023-01-28', '2023-01-29', '2023-01-30', 
    '2023-01-31', '2023-02-01'
]
prices = [
    150, 152, 148, 145, 149, 151, 155, 156, 157, 158,
    160, 162, 161, 163, 165, 168, 169, 170, 172, 171,
    174, 175, 176, 178, 177, 180, 181, 182, 183, 185,
    184, 186
]

# Convert lists to a DataFrame
df = pd.DataFrame({
    'Date': pd.to_datetime(dates),  # Convert dates to datetime format
    'Price': prices
})

# Calculate the SMA and LMA
df['SMA'] = df['Price'].rolling(window = 5).mean()
df['LMA'] = df['Price'].rolling(window = 20).mean()

# Display the resulting DataFrame
print(df.tail(5))

# Plot
def plot_stock_with_moving_averages(df):
    # Plot stock prices
    plt.plot(df.index, df['Price'], label='Stock Prices', color='blue')
    
    # Plot 5-day moving average
    plt.plot(df.index, df['SMA'], label='5-Day SMA', color='green')

    # Plot 20-day moving average
    plt.plot(df.index, df['LMA'], label='20-Day SMA', color='red')

    # Customize the plot
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Stock Price and Moving Averages')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function to plot
plot_stock_with_moving_averages(df)
```
